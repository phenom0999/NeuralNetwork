## MNIST Dataset: Handwritten Digits 

```python
import pandas as pd # type: ignore
import matplotlib.pyplot as plt # type: ignore
import numpy as np 
from NeuralNetwork import NeuralNetwork
from Matrix import Matrix



df = pd.read_csv("data/mnist_train.csv")

df.head()
```

```python
# Select 36 rows (images) from the dataset
num_images = 100
indices = range(1000, 1000 + num_images)  # Adjust starting index as needed

fig, axes = plt.subplots(10, 10, figsize=(10, 10))

for i, ax in enumerate(axes.flat):
    # Extract the row
    row = df.iloc[indices[i]].values

    # Get the label and pixel values
    label = row[0]
    pixels = row[1:].reshape(28, 28)

    # Display the image
    ax.imshow(pixels, cmap='gray')
    ax.set_title(f"Label: {label}", fontsize=8)
    ax.axis("off")  # Remove axis for a cleaner look

plt.tight_layout()
plt.show()
```
